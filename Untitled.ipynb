```python
# Balanced Extraction on synthetic SWOT KaRIn + Nadir data
# saves ht_all and posterior covariance matrix P

import pickle
import numpy as np
import xarray as xr
import scipy.linalg as la
from scipy.linalg import block_diag
import JWS_SWOT_toolbox as swot
t = swot.Timer()

# --- Load synthetic data ---
karin_NA = pickle.load(open("./pickles/karin_NA.pkl", "rb"))
nadir_NA = pickle.load(open("./pickles/nadir_NA.pkl", "rb"))
karin    = pickle.load(open("./pickles/karin.pkl", "rb"))
nadir    = pickle.load(open("./pickles/nadir.pkl", "rb"))
t.lap("Data loaded")

# --- Fit spectra (work in cm) ---
ntime = karin_NA.ssha.shape[0]

ssh_noisy_xr = xr.DataArray(
    karin_NA.ssh_noisy * 100.0,
    coords=[np.arange(ntime), karin_NA.y_coord_km, karin_NA.x_coord_km],
    dims=["sample", "line", "pixel"],
)
spec_ssh_noisy = swot.mean_power_spectrum(
    ssh_noisy_xr, karin_NA.window, "line", ["sample", "pixel"]
)

nad_noisy_xr = xr.DataArray(
    nadir_NA.ssh_noisy * 100.0,
    coords=[np.arange(ntime), nadir_NA.y_coord_km],
    dims=["sample", "nadir_line"],
)
spec_nad_noisy = swot.mean_power_spectrum(
    nad_noisy_xr, nadir_NA.window, "nadir_line", ["sample"]
)

poptcwg_karin_NA, _ = swot.fit_spectrum(karin_NA, spec_ssh_noisy, swot.karin_model)
poptcwg_nadir_NA, _ = swot.fit_nadir_spectrum(nadir_NA, spec_nad_noisy, poptcwg_karin_NA)
swot.plot_spectral_fits(karin_NA, nadir_NA, poptcwg_karin_NA, poptcwg_nadir_NA,"spec_fits.pdf")
t.lap("Spectral fitting done")
```

```python
# --- Balanced Extraction setup ---
dn = nadir.dy_km
#c   = swot.balanced_covariance_func(poptcwg_karin_NA)
c   = swot.balanced_covariance_func(poptcwg_karin_NA, smooth=True, scale_km=32) 
nk  = swot.noise_covariance_func(poptcwg_karin_NA)
sigma_n = np.sqrt(poptcwg_nadir_NA[0] / (2 * dn))  # nadir white-noise std (cm)

idx = 40
mask_k = ~np.isnan(karin_NA.ssha[idx])
mask_n = ~np.isnan(nadir_NA.ssha[idx])

xkk = karin_NA.x_grid[mask_k].ravel(order="C") * 1e-3
ykk = karin_NA.y_grid[mask_k].ravel(order="C") * 1e-3

xobs = np.concatenate([xkk])
yobs = np.concatenate([ykk])
n_k = xkk.size

dx_k = xobs[:n_k, None] - xobs[:n_k]
dy_k = yobs[:n_k, None] - yobs[:n_k]
Nk_obs = nk(np.hypot(dx_k, dy_k)) * 0.0
N_obs = block_diag(Nk_obs)

C_obs = swot.build_covariance_matrix(c, xobs, yobs)

jitter = 1e-8 * np.nanmedian(np.diag(C_obs))
A_spd  = 0.5 * (C_obs + N_obs) + jitter * np.eye(C_obs.shape[0])

#cho_tuple = la.cho_factor(C_obs + N_obs, lower=True) # Cholesky factor (C + N)^-1
cho_tuple = la.cho_factor(A_spd, lower=True) # Cholesky factor (C + N)^-1

xt, yt, nxt, nyt, _, _ = swot.make_target_grid(karin_NA, unit="km", extend=False)
R = c(np.hypot(xt[:, None] - xobs, yt[:, None] - yobs))
t.lap("Setup complete")

# --- Run extraction across time ---
print("Running balanced extraction on synthetic KaRIn + Nadir data...")
n_times = karin_NA.ssh_noisy.shape[0]
ht_all = np.empty((n_times, nyt, nxt), dtype=float)

for i in range(n_times):
    h_obs = np.concatenate([
        100.0 * karin_NA.ssha_full[i, :, 4:64][mask_k].ravel(order="C")
    ])
    ht = swot.estimate_signal_on_target_cho_solve(R, cho_tuple, h_obs, output=False)
    ht_all[i] = (ht / 100.0).reshape(nyt, nxt)
    if (i + 1) % 10 == 0 or i == n_times - 1:
        print(f"  processed {i + 1}/{n_times}")

t.lap("Balanced extraction finished")

# --- Save balanced field ---
with open("./pickles/balanced_extraction_synth_NA_nonoise_sm_32km.pkl", "wb") as f:
    pickle.dump(ht_all, f)
t.lap("Saved balanced field")

t.total()
```

```python
# Simple: plot ζ/f PDFs at t=0 for 2, 4, 8, 16 km
# Overlays "nonoise" vs "with noise" for each scale.

import os, pickle
import numpy as np
import matplotlib.pyplot as plt
import JWS_SWOT_toolbox as swot

# --- config ---
PICKLES = "./pickles"
KARIN_NA_PATH = f"{PICKLES}/karin_NA.pkl"
SCALES = [1, 2, 4, 8, 16]   # km
TIDX = 10                 # time = 0

def path_nonoise(km): return f"{PICKLES}/balanced_extraction_synth_NA_tmean_nonoise_sm_{km}km.pkl"
def path_withnoise(km): return f"{PICKLES}/balanced_extraction_synth_NA_tmean_sm_{km}km.pkl"

def load(p):
    with open(p, "rb") as f:
        return pickle.load(f)

# --- grid / latitude once ---
karin_NA = load(KARIN_NA_PATH)
dx_m = float(karin_NA.dx_km) * 1e3
dy_m = float(karin_NA.dy_km) * 1e3

# latitude as 1D (y)
lats = np.asarray(getattr(karin_NA, "lat"), dtype=float)
lat_1d = lats[0, :, 0] if lats.ndim == 3 else (lats[:, 0] if lats.ndim == 2 else lats)

print(lats.shape)
print(lat_1d.shape)

# --- plotting ---
fig, axes = plt.subplots(2, 3, figsize=(9, 7), sharex=True, sharey=True)
axes = axes.ravel()

for ax, km in zip(axes, SCALES):
    # load SSH fields [time, y, x], pick t=0
    ht_no = np.asarray(load(path_nonoise(km))[TIDX], dtype=float)
    ht_wi = np.asarray(load(path_withnoise(km))[TIDX], dtype=float)

    # geostrophic vorticity at t=0
    z_no = swot.compute_geostrophic_vorticity_5pt(ht_no, dx_m, dy_m, lat_1d)
    z_wi = swot.compute_geostrophic_vorticity_5pt(ht_wi, dx_m, dy_m, lat_1d)

    # common bins (robust) and PDFs
    a = z_no[np.isfinite(z_no)].ravel()
    b = z_wi[np.isfinite(z_wi)].ravel()
    lo = np.percentile(np.concatenate([a, b]), 0.5)
    hi = np.percentile(np.concatenate([a, b]), 99.5)
    edges = np.linspace(lo, hi, 121)
    centers = 0.5 * (edges[:-1] + edges[1:])
    H_no, _ = np.histogram(a, bins=edges, density=True)
    H_wi, _ = np.histogram(b, bins=edges, density=True)

    # plot
    ax.plot(centers, H_no, lw=2, label="NA sim. ζ/f")
    ax.plot(centers, H_wi, lw=2, label="Extracted ζ/f")
    ax.set_title(f"{km} km smoothing")
    ax.grid(True, ls=":", lw=0.7)

# labels & legend
axes[2].set_xlabel("Geostrophic vorticity ζ/f")
axes[3].set_xlabel("Geostrophic vorticity ζ/f")
axes[0].set_ylabel("Probability density")
axes[2].set_ylabel("Probability density")

handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, ncol=2, loc="upper center", frameon=False)
#fig.suptitle("t = 0 PDFs of ζ/f (no-noise vs with-noise)", y=0.98)
fig.tight_layout(rect=[0, 0, 1, 0.95])

# save if you like:
# fig.savefig("zeta_pdfs_t0_2x2.pdf", dpi=300)
plt.show()
```

```python
import pickle, numpy as np, matplotlib.pyplot as plt, scipy.linalg as la
from scipy.ndimage import gaussian_filter
import JWS_SWOT_toolbox as swot

PICKLES = "./pickles"
POST = f"{PICKLES}/posterior_full_synth_NA_tmean_sm_8km.pkl"
SSH  = f"{PICKLES}/balanced_extraction_synth_NA_tmean_sm_4km.pkl"
KARIN= f"{PICKLES}/karin_NA.pkl"

# --- Load files ---
C   = pickle.load(open(POST, "rb"))
ht  = pickle.load(open(SSH,  "rb"))  # (t, y, x)
karin = pickle.load(open(KARIN, "rb"))

dx, dy = karin.dx_km * 1e3, karin.dy_km * 1e3
lat = karin.lat[0, :, 0]
```

```python
# --- Make C symmetric, then zero out all negative entries ---
C = 0.5 * (C + C.T)

# Cholesky
#L = la.cholesky(C + np.eye(C.shape[0])*1e-3, lower=True)
eigvals, eigvecs = la.eigh(C)
eigvals = np.maximum(eigvals, 1e-10)
L = eigvecs @ np.diag(np.sqrt(eigvals))

# --- Draw posterior sample(s) ---
num_samples = 1
times = ht.shape[0]
ny, nx = ht.shape[1:]
all_zeta = np.empty((times, num_samples, ny, nx))

for t in range(times):
    mu = ht[t].ravel() * 100.0  # cm
    print(t)
    for s in range(num_samples):
        Z = np.random.randn(mu.size)
        ssh_flat = (mu + L @ Z) / 100.0  # meters
        ssh = ssh_flat.reshape(ny, nx)
        zeta = swot.compute_geostrophic_vorticity(ssh, dx, dy, lat)
        all_zeta[t, s] = zeta

# --- Plot one field ---
plt.imshow(all_zeta[0, 0], origin="lower", cmap="RdBu_r")
plt.colorbar(label="ζ / f")
plt.title("Posterior sample vorticity")
plt.tight_layout()
plt.show()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
from scipy.stats.mstats import winsorize
from scipy.ndimage import gaussian_filter
import scipy.linalg as la
import JWS_SWOT_toolbox as swot

# =========================
#  --- CONFIG ---
# =========================
PICKLES = "./pickles"
KARIN = f"{PICKLES}/karin_NA.pkl"
POST = f"{PICKLES}/posterior_full_synth_NA.pkl"
SCALES = [1, 2, 4, 8, 16]  # km
FINAL = f"{PICKLES}/final_plot_data.pkl"

def path_nonoise(s): return f"{PICKLES}/balanced_extraction_synth_NA_tmean_nonoise_sm_{s}km.pkl"
def path_withnoise(s): return f"{PICKLES}/balanced_extraction_synth_NA_tmean_sm_{s}km.pkl"

# =========================
#  --- LOAD GRID ---
# =========================
karin = pickle.load(open(KARIN, "rb"))
dx_m, dy_m = karin.dx_km * 1e3, karin.dy_km * 1e3
lat_1d = karin.lat[0, :, 0]

# =========================
#  --- HELPERS ---
# =========================
def finite_flat(a): return np.asarray(a).ravel()[np.isfinite(a)]

def stack_vort(ssh_stack):
    lats_sliced = lat_1d[:ssh_stack.shape[1]]
    return np.stack([swot.compute_geostrophic_vorticity(a, dx_m, dy_m, lats_sliced)
                     for a in ssh_stack], axis=0)

def _sigma_pixels(scale_km):
    σ = scale_km / (2 * np.pi)
    return (σ * 1e3 / dy_m, σ * 1e3 / dx_m)

def posterior_draw_vort_samples(mean_ssh, L, scale_km, n_samples=1, rng=None):
    """Draw posterior SSH samples → smooth → ζ/f."""
    if rng is None: rng = np.random.default_rng(0)
    nY, nX = mean_ssh.shape
    mu_cm = mean_ssh.ravel() * 100.0
    Z = rng.standard_normal((mu_cm.size, n_samples))
    ssh = (mu_cm[:, None] + L @ Z).T.reshape(n_samples, nY, nX) / 100.0
    sy, sx = _sigma_pixels(scale_km)
    ssh_sm = gaussian_filter(ssh, sigma=(0, sy, sx), mode="nearest", truncate=50.0)
    vort = stack_vort(ssh_sm)
    return finite_flat(vort), vort[0]

def compute_stats_from_stack(z_stack, edges):
    """Compute PDFs + low-order stats over time."""
    centers = 0.5 * (edges[:-1] + edges[1:])
    pdfs = np.zeros((z_stack.shape[0], centers.size))
    stds = vars_ = skews = kurts = []
    for i, z in enumerate(z_stack):
        vals = finite_flat(z)
        if vals.size < 2: continue
        h, _ = np.histogram(vals, bins=edges, density=True)
        pdfs[i] = h
        vals = winsorize(vals, limits=[0.001, 0.001]).data
        stds.append(np.std(vals))
        vars_.append(np.var(vals))
        skews.append(skew(vals))
        kurts.append(kurtosis(vals, fisher=True))
    return (
        (centers, np.nanmean(pdfs, 0), np.nanstd(pdfs, 0)),
        dict(
            var_mu=np.nanmean(vars_), std_mu=np.nanmean(stds),
            skw_mu=np.nanmean(skews), kur_mu=np.nanmean(kurts)
        ),
    )

# =========================
#  --- LOAD POSTERIOR ---
# =========================
P = pickle.load(open(POST, "rb"))
P = 0.5 * (P + P.T)
L_post = la.cholesky(P + np.eye(P.shape[0]) * 1e-8, lower=True)

# =========================
#  --- MAIN LOOP ---
# =========================
stats, all_pdfs = {"scale": []}, []
for k in ["var_no_mu","std_no_mu","skw_no_mu","kur_no_mu",
          "var_wi_mu","std_wi_mu","skw_wi_mu","kur_wi_mu",
          "var_post","std_post","skw_post","kur_post"]:
    stats[k] = []

for scale in SCALES:
    print(f"\nScale {scale} km")

    ht_no = np.asarray(pickle.load(open(path_nonoise(scale), "rb")), float)
    ht_wi = np.asarray(pickle.load(open(path_withnoise(scale), "rb")), float)
    z_no, z_wi = stack_vort(ht_no), stack_vort(ht_wi)
    edges = np.linspace(-2, 2, 121)

    (_, m_no, _), tm_no = compute_stats_from_stack(z_no, edges)
    (_, m_wi, _), tm_wi = compute_stats_from_stack(z_wi, edges)

    # Posterior samples
    all_h, all_v, all_s, all_sk, all_k = [], [], [], [], []
    for t in range(ht_wi.shape[0]):
        vals, _ = posterior_draw_vort_samples(ht_wi[t], L_post, scale, n_samples=20)
        h, _ = np.histogram(vals, bins=edges, density=True)
        all_h.append(h)
        v = winsorize(vals, [0.001, 0.001]).data
        all_v.append(np.var(v)); all_s.append(np.std(v))
        all_sk.append(skew(v)); all_k.append(kurtosis(v, fisher=True))
    m_post = np.nanmean(all_h, 0)

    # Store
    stats["scale"].append(scale)
    for k, d in [("no", tm_no), ("wi", tm_wi)]:
        for f in ["var","std","skw","kur"]:
            stats[f"{f}_{k}_mu"].append(d[f"{f}_mu"])
    stats["var_post"].append(np.nanmean(all_v))
    stats["std_post"].append(np.nanmean(all_s))
    stats["skw_post"].append(np.nanmean(all_sk))
    stats["kur_post"].append(np.nanmean(all_k))

    all_pdfs.append({"scale": scale, "x": 0.5*(edges[1:]+edges[:-1]),
                     "m_no": m_no, "m_wi": m_wi, "m_post": m_post})

pickle.dump({"stats": stats, "all_pdf_data": all_pdfs}, open(FINAL, "wb"))

# =========================
#  --- FIGURE 4: Example ζ fields ---
# =========================
print("\nGenerating example vorticity fields...")
dx_km, dy_km = dx_m / 1e3, dy_m / 1e3
t0 = 0
nrows, ncols = len(SCALES), 3
fig, axs = plt.subplots(nrows, ncols, figsize=(12, 2.6*nrows))
plt.subplots_adjust(hspace=0.3, wspace=0.15)

for r, scale in enumerate(SCALES):
    ht_no = np.asarray(pickle.load(open(path_nonoise(scale), "rb")), float)
    ht_wi = np.asarray(pickle.load(open(path_withnoise(scale), "rb")), float)
    z_no, z_wi = stack_vort(ht_no[t0:t0+1])[0], stack_vort(ht_wi[t0:t0+1])[0]
    _, z_po = posterior_draw_vort_samples(ht_wi[t0], L_post, scale, n_samples=1)

    pools = np.concatenate([finite_flat(z_no), finite_flat(z_wi), finite_flat(z_po)])
    vmax = np.nanpercentile(np.abs(pools), 99)
    extent = [0, z_no.shape[1]*dx_km, 0, z_no.shape[0]*dy_km]

    for c, (title, field) in enumerate(zip(["NA sim.", "Extracted", "Posterior"], [z_no, z_wi, z_po])):
        ax = axs[r, c] if nrows > 1 else axs[c]
        im = ax.imshow(field, cmap="RdBu_r", vmin=-vmax, vmax=vmax,
                       origin="lower", extent=extent, aspect="auto")
        if r == 0: ax.set_title(f"{title} ζ", fontsize=13)
        if c == 0: ax.set_ylabel(f"{scale} km\n y [km]", fontsize=11)
        if r == nrows-1: ax.set_xlabel("x [km]", fontsize=11)
        ax.tick_params(labelsize=10)
        if c == ncols-1: last_im = im

fig.colorbar(last_im, ax=axs, location="right", fraction=0.025, pad=0.04, label="ζ / f")
fig.savefig("Vorticity_Example_Fields_Simple.pdf", dpi=200, bbox_inches="tight")
plt.close(fig)
print("Saved Vorticity_Example_Fields_Simple.pdf")
```
